```python
from helper import plot_tree_for_jupyter, data_manager
from sklearn.tree import DecisionTreeRegressor
from IPython.display import HTML
####### 定数宣言　#######
# csvファイルのパス
CSV_PATH = "./data/train.csv"
# 目的変数名
OBJ_VAL = "SalePrice"
# ダミー化する変数名
DUMMY_VAL = None          
```

```python
# 分析したい変数を目的変数と従属変数に分割
x, y, X, Y = data_manager.get_data(CSV_PATH, OBJ_VAL, DUMMY_VAL)
```

```python
# sklearn の回帰決定木で分析
# サンプル数が極端に少ない葉が出てしまう時は、min_samples_splitやmin_samples_leafで調整する。
clf = regressor = DecisionTreeRegressor(max_depth=5, min_samples_split=20, min_samples_leaf=10)
clf.fit(x, y)
```

```python
# 分析結果をプロット
HTML(plot_tree_for_jupyter.plot_tree(x, y, clf, OBJ_VAL, 20))
```
